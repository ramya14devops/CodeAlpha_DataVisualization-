```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("supermarket_data.csv")
print(df.head())
```

```python
print(df.columns)
```

```python
df['Sales']=df['Quantity']*df['Price']
print(df.head())
```

```python
country_sales=df.groupby("Country")["Sales"].sum().sort_values(ascending=False).head(10)
country_sales.plot(kind="bar")
plt.title("Top 10 countries by Sales")
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.show()
```

```python
country_sales=df.groupby("Country")["Sales"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,8))
plt.pie(country_sales,labels=country_sales.index,autopct='%1.1f%%')
plt.title("Sales Distribution by top 5 countries")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("supermarket_data.csv")
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
print(df[["InvoiceDate", "Month"]].head())
```

```python
df['Sales']=df['Quantity']*df['Price']
print(df.head())
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()

monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_products = df.groupby("Description")["Sales"].sum().sort_values(ascending=False).head(10)

top_products.plot(kind="bar")

plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

Key Findings

1.The top 10 countries contribute the majority of total sales.
2.The top 5 countries account for a significant percentage of overall revenue.
3.Monthly sales trends reveal fluctuations in customer demand over time.
4.Certain products generate substantially higher sales compared to others.
5.Sales data can help improve inventory and marketing decisions.
